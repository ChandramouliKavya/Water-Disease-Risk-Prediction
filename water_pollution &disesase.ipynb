Access to clean and safe water is a pressing global issue, especially in developing and densely populated regions where water pollution and poor sanitation contribute significantly to the spread of waterborne diseases like cholera, typhoid, and diarrheal infections. Government agencies, non-profit organizations, and environmental consultancies often lack the tools to proactively monitor and mitigate these risks.

The challenge lies in predicting which areas are most at risk due to fluctuating water quality indicators (e.g., pH level, turbidity, nitrate concentration), socioeconomic factors (e.g., healthcare access, GDP per capita), and environmental variables (e.g., rainfall, temperature). Currently, decisions are reactive, leading to delayed intervention and increased healthcare costs.



# **Data Description** 

**water_source_type: The type of water source (e.g., well, river, piped, rainwater).**

**contaminant_level_(ppm): Total concentration of harmful substances in water measured in parts per million.**

**ph_level: Acidity or alkalinity of water on the pH scale (ideal drinking water: ~6.5–8.5).**

**turbidity_(ntu): Cloudiness or haziness of water caused by particles, measured in NTU (Nephelometric Turbidity Units).**

**dissolved_oxygen_(mg/l): Amount of oxygen dissolved in water, indicating water quality (important for aquatic life).**

**nitrate_level_(mg/l): Nitrate concentration; high levels can indicate pollution and cause health issues like methemoglobinemia**.

**lead_concentration_(µg/l): Lead content in water, measured in micrograms per liter. Lead is toxic and should be minimal.**

**bacteria_count_(cfu/ml): Number of bacterial colonies per milliliter, indicating biological contamination.**

**water_treatment_method: Type of treatment applied to make water safe (e.g., chlorination, boiling, filtration).**

**access_to_clean_water_(%_of_population): Percentage of the population with access to safe, clean drinking water.**

**diarrheal_cases_per_100,000_people: Reported cases of diarrhea per 100,000 people — commonly linked to unsafe water.**

**cholera_cases_per_100,000_people: Reported cholera cases per 100,000 people.**

**typhoid_cases_per_100,000_people: Reported typhoid cases per 100,000 people.**

**infant_mortality_rate_(per_1,000_live_births): Number of infants dying before age 1 per 1,000 live births, often tied to waterborne diseases.**

**gdp_per_capita_(usd): Economic indicator measuring average income per person in USD.**

**healthcare_access_index_(0-100): Index from 0 to 100 showing access to healthcare; higher values indicate better access.**

**urbanization_rate_(%): Percentage of the population living in urban areas.**

**sanitation_coverage_(%_of_population): Percentage of the population with access to proper sanitation facilities.**

**rainfall_(mm_per_year): Annual rainfall in millimeters, influencing water availability and quality.**

**temperature_(°c): Average ambient temperature in degrees Celsius — affects bacterial growth and waterborne disease spread.**

**water_safety: A classification label indicating if the water is safe or unsafe (binary or multiclass).**

**disease_risk_level: A label showing the overall risk level of water-related diseases (e.g., low, medium, high).**

```python
# Function to create and train pipeline
def run_regression_pipeline(df, target_col, model):
    # 1. Separate features and target
    X = df.drop(columns=[target_col])
    y = df[target_col]

    # 2. Identify column types
    num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

    # 3. Define transformers
    numerical_pipeline = Pipeline([('scaler', StandardScaler()) ])

    categorical_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    # 4. Combine transformers
    preprocess = ColumnTransformer([
        ('numerical_process', numerical_pipeline, num_cols),
        ('categorical_process', categorical_pipeline, cat_cols)
    ])

    # 5. Final pipeline
    pipeline = Pipeline([
        ('preprocessor', preprocess),
        ('regressor', model)
    ])

    # 6. Train-test split
    from sklearn.model_selection import train_test_split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)

    # 7. Fit and predict
    pipeline.fit(X_train, y_train)
    y_train_pred = pipeline.predict(X_train)
    y_test_pred = pipeline.predict(X_test)

    # 8. Evaluation
    print(f"\nModel: {model.__class__.__name__}")
    print("Train R²:", r2_score(y_train, y_train_pred))
    print("Test R²:", r2_score(y_test, y_test_pred))
    print("="* 60)
    print("Train RMSE:", mean_squared_error(y_train, y_train_pred, squared=False))
    print("Test RMSE:", mean_squared_error(y_test, y_test_pred, squared=False))

    return pipeline

# Example usage (replace with your dataframe and model choice):
# model = LinearRegression()
# pipeline = run_regression_pipeline(df, 'diarrheal cases per 100,000 people', model)

# Save to pickle
# with open('regression_model.pkl', 'wb') as f:
#     pickle.dump(pipeline, f)
```

```python
def run_classification_pipeline(df, target_col, model, use_smote = True):
    # 1. Separate features and target
    X = df.drop(columns=[target_col])
    y = df[target_col]

    # 2. Identify column types
    num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

    
    ordinal_cols = ['water_safety'] if 'water_safety' in cat_cols else []
    nominal_cols = list(set(cat_cols) - set(ordinal_cols))

    numerical_pipeline = Pipeline([
        ('scaler', StandardScaler())
    ])

    nominal_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    ordinal_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinal', OrdinalEncoder())
    ])

    # 4. Combine transformers
    preprocess = ColumnTransformer([
        ('num', numerical_pipeline, num_cols),
        ('nom', nominal_pipeline, nominal_cols),
        ('ord', ordinal_pipeline, ordinal_cols)
    ])

    # 5. Final pipeline with SMOTE
    if use_smote:
        pipeline = ImbPipeline([
            ('preprocessor', preprocess),
            ('smote', SMOTE(random_state=42)),
            ('classifier', model)
        ])
    else:
        pipeline = Pipeline([
            ('preprocessor', preprocess),
            ('classifier', model)
        ])

    # 6. Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27)

    # 7. Fit and Predict
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    # 8. Evaluation
    print("\n📊 Classification Metrics:")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print(f"Recall:    {recall_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print(f"F1 Score:  {f1_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))

    return pipeline
```

```python
# Function to create and train classification pipeline with ordinal handling and SMOTE
def run_classification_pipeline(df, target_col, model, use_smote = True):
    # 1. Separate features and target
    X = df.drop(columns=[target_col])
    y = df[target_col]

    #  Identify column types
    num_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
    cat_cols = X.select_dtypes(include=['object', 'category']).columns.tolist()

    #  Manually define ordinal and nominal columns
    ordinal_cols = ['disease_risk_level'] if 'disease_risk_level' in cat_cols else []
    nominal_cols = list(set(cat_cols) - set(ordinal_cols))

    #  Define individual pipelines
    numerical_pipeline = Pipeline([
        ('scaler', StandardScaler())
    ])

    nominal_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(handle_unknown='ignore'))
    ])

    ordinal_pipeline = Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('ordinal', OrdinalEncoder())
    ])

    #  Combine all preprocessing steps
    preprocess = ColumnTransformer([
        ('num', numerical_pipeline, num_cols),
        ('nom', nominal_pipeline, nominal_cols),
        ('ord', ordinal_pipeline, ordinal_cols)
    ])

    #  Final pipeline with SMOTE and classifier
    if use_smote:
        pipeline = ImbPipeline([
            ('preprocessor', preprocess),
            ('smote', SMOTE(random_state=27)),
            ('classifier', model)
        ])
    else:
        pipeline = Pipeline([
            ('preprocessor', preprocess),
            ('classifier', model)
        ])

    # Train-test split
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=27, stratify=y)

    # Fit and Predict
    pipeline.fit(X_train, y_train)
    y_pred = pipeline.predict(X_test)

    #  Evaluation
    print("\n📊 Classification Metrics:")
    print(f"Accuracy:  {accuracy_score(y_test, y_pred):.4f}")
    print(f"Precision: {precision_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print(f"Recall:    {recall_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print(f"F1 Score:  {f1_score(y_test, y_pred, average='weighted', zero_division=0):.4f}")
    print("\nClassification Report:\n", classification_report(y_test, y_pred, zero_division=0))

    return pipeline
```
